```python
import sys
!{sys.executable} -m pip install python-docx

# ==========================================
# FORWARD PROPAGATION (REGRESSION NETWORK)
# ==========================================

# Import library to create Google Docs–compatible file
from docx import Document

# --------------------
# Given Inputs
# --------------------
a = 20
b = 30
c = 15

# Weights
w1 = 0.6
w2 = 0.3
w3 = 0.6
w4 = 0.3

# --------------------
# Hidden Layer Calculations
# --------------------
h1 = (a * w1) + (b * w2) + (c * w1)
h2 = (a * w3) + (b * w4) + (c * w3)

# --------------------
# Output Layer (Regression – Linear Activation)
# --------------------
y = (h1 * w1) + (h2 * w2)

# --------------------
# Create Google Docs File
# --------------------
doc = Document()
doc.add_heading("Forward Propagation Calculation", level=1)

doc.add_paragraph("Given Values:")
doc.add_paragraph(f"a = {a}, b = {b}, c = {c}")
doc.add_paragraph(f"w₁ = w₃ = {w1}")
doc.add_paragraph(f"w₂ = w₄ = {w2}")

doc.add_heading("Hidden Layer Computation", level=2)
doc.add_paragraph(f"h₁ = (20 × 0.6) + (30 × 0.3) + (15 × 0.6) = {h1}")
doc.add_paragraph(f"h₂ = (20 × 0.6) + (30 × 0.3) + (15 × 0.6) = {h2}")

doc.add_heading("Output Layer Computation", level=2)
doc.add_paragraph("Regression problem → Linear activation function")
doc.add_paragraph(f"y = (h₁ × 0.6) + (h₂ × 0.3)")
doc.add_paragraph(f"y = {y}")

# Save file
file_path = "/content/Forward_Propagation_Calculation.docx"
doc.save(file_path)

file_path
```
